```python
import numpy
import matplotlib.pyplot #import library
data = numpy.loadtxt('inflammation-01.csv', delimiter = ",")

#Create a figure 1 row, 3 columns
fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

plot1 = fig.add_subplot(1,3,1) #here are three figures
plot2 = fig.add_subplot(1,3,2) #how many rows, how many columns, which is it
plot3 = fig.add_subplot(1,3,3)
plot1.set_ylabel('Average') #give each figure a label
plot1.plot(numpy.mean(data, axis=0))

plot2.set_ylabel('Max')
plot2.set_ylim(0,21)
plot2.plot(numpy.max(data, axis=0))

plot3.set_ylabel('Min')
plot3.set_ylim(0,5.5)
plot3.plot(numpy.min(data, axis=0))

fig.tight_layout() #spaces them out, yes that makes no sense
matplotlib.pyplot.show()
```

```python
import numpy
import matplotlib.pyplot #import library
data = numpy.loadtxt('inflammation-01.csv', delimiter = ",")

#Create a figure 1 row, 3 columns
fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

plot1 = fig.add_subplot(1,4,1) #here are three figures
plot2 = fig.add_subplot(1,4,2) #how many rows, how many columns, which is it
plot3 = fig.add_subplot(1,4,3)
plot4 = fig.add_subplot(1,4,4)

plot1.set_ylabel('Average') #give each figure a label
plot1.plot(numpy.mean(data, axis=0))

plot2.set_ylabel('Max')
plot2.set_ylim(0,21)
plot2.plot(numpy.max(data, axis=0))

plot3.set_ylabel('Min')
plot3.set_ylim(0,5.5)
plot3.plot(numpy.min(data, axis=0))

plot4.set_ylabel('Stdev')
plot4.plot(numpy.std(data, axis=0))

fig.tight_layout() #spaces them out, yes that makes no sense
matplotlib.pyplot.show()
```

```python
#grab list of files from directory you are in, 
#function called glob that finds names and directories that match a pattern
#glob.glob - library.function
import glob
print (glob.glob('inflammation*.csv'))
```

```python
import numpy   #Don't have to keep reimporting libraries you have already imported
import matplotlib.pyplot
import glob

filenames = sorted(glob.glob('inflammation*.csv'))  #Sorts list
filenames = filenames[0:3]   #Just want first three from sorted list

for f in filenames:
    print(f)  #print filename
    
    data = numpy.loadtxt(fname = f, delimiter=',')   #read data into dumpy file, don't need in quotes because in list
    
    fig = matplotlib.pyplot.figure(figsize=(10.0,3.0))
    
    plot1=fig.add_subplot(1,3,1)
    plot2=fig.add_subplot(1,3,2)
    plot3=fig.add_subplot(1,3,3)
    
    plot1.set_ylabel('average')
    plot1.plot(numpy.mean(data,axis=0))
    
    plot2.set_ylabel('max')
    plot2.plot(numpy.max(data,axis=0))

    plot3.set_ylabel('min')
    plot3.plot(numpy.min(data,axis=0))
    
    fig.tight_layout()
    matplotlib.pyplot.show()
```
